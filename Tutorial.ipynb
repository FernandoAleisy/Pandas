```python
import pandas as pd
personas=pd.read_csv('Enero\Área - Características generales (Personas).csv', sep=';')
personas
```

```python
print(
list(personas.columns)
)
```

```python
personas['fex_c_2011']=personas['fex_c_2011'].str.replace(',','.')
personas['fex_c_2011']=pd.to_numeric(personas['fex_c_2011'], errors='coerce')
personas['fex_c_2011']=personas['fex_c_2011'].round().astype(int)
print(personas.dtypes)
```

```python
for i in personas:
    if personas[i].dtypes==object:
        personas[i]=pd.to_numeric(personas[i], errors='coerce')
personas.dtypes
```

```python
personas.head(10)
```

Identificar las preguntas que no fueron respondidas

```python
for i in personas:
    if personas[i].count()==0:
        print(i)
```
