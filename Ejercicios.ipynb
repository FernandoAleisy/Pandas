<b style=color:#01befe>Escribir un programa que pregunte al usuario por las ventas de un rango de años y muestre por pantalla una serie con los datos de las ventas indexada por los años, antes y después de aplicarles un descuento del 10%.</b>

```python
import pandas as pd

def IngresoDatos():
    Estado=True
    diccionario={'año':[],'ventas':[],'VDescuentos':[],'Satisfacción':[]}
    while Estado==True:
        diccionario['año'].append(input('Ingrese el año'))
        ventas=float(input('¿Cuál fue su facturación anual?'))
        diccionario['ventas'].append(ventas)
        diccionario['VDescuentos'].append(ventas*0.9)
        diccionario['Satisfacción'].append('Satisfecho')
        pregunta=input('¿Deseas ingresar los datos de otro año?')
        if pregunta.upper()=='NO':
            Estado=False
    return pd.DataFrame(diccionario)
    
MisDatos=IngresoDatos()
MisDatos
```

<b style=color:#01befe>Escribir una función que reciba un diccionario con las notas de los alumno de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.</b>

```python
import statistics as sta
import pandas as pd

def estadisticas(diccionario):
    if type(diccionario)==dict:
        resultado={'Nombre':[],'NMinima':[],'NMaxima':[],'Media':[],'DesvTip':[]}
        for i in diccionario:
            resultado['Nombre'].append(i)
            resultado['NMinima'].append(min(diccionario[i]))
            resultado['NMaxima'].append(max(diccionario[i]))
            resultado['Media'].append(sta.mean(diccionario[i]))
            resultado['DesvTip'].append(sta.pstdev(diccionario[i]))
        return resultado

Estudiantes={'Carlos':[4.2,5.0,5.0,4.4,3.8],'Sandra':[4.0,5.0,4.7,4.8,4.3]}
datos=pd.DataFrame(estadisticas(Estudiantes))
datos
```

<b style=color:#01befe>Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con las notas de los alumnos aprobados ordenadas de mayor a menor.</b>

```python
import pandas as pd

def pasan(diccionario):
    clave=[i for i in diccionario]
    Estudiantes,Notas=diccionario[clave[0]],diccionario[clave[1]]
    union=list(zip(Notas,Estudiantes))
    union.sort(reverse=True)
    union=[(i,j) for i,j in union if i>=3.0]
    diccionario={'Estudiante':[j for _,j in union],'Notas':[i for i,_ in union]}
    return pd.DataFrame(diccionario)

Notas={'Estudiante':['Felipe','Silvina','Lucas','Camilo'],'Nota':[4.0,3.8,4.2,2.5]}
Pasaron=pasan(Notas)
Pasaron
```
