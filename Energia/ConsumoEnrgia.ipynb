```python
resultado=datos[datos['Year']==2019]
paises=['Argentina','Brazil','Chile','Colombia','Ecuador','Mexico','Peru']
ind=[]
for i,j in enumerate(resultado['Entity']):
    if j in paises:
        ind.append(i)
ind

resultado=resultado.iloc[ind]
resultado
```

```python
maximo=resultado['Wind_Generation_TWh'].max()
pais=resultado[resultado['Wind_Generation_TWh']==maximo]
final=pd.DataFrame(pais.groupby(['Entity']))
#final=pd.DataFrame(pais.groupby(['Entity','Wind_Generation_TWh']))
p=str(final[0].to_numpy())[2:-2]
p
```
